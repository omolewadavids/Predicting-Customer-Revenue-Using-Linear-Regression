# Prediction Customer Revenue Using Linear Regression

### Import pandas and read the dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('retail_transaction.csv')
```

```python
df.head()
```

```python
len(df)
```

### Data Wrangling

```python
# Convert the InvoiceDate column to date format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df
```

#### Calculate revenue for each row 

```python
df['revenue'] = df['UnitPrice'] * df['Quantity']
```

```python
operations = {'revenue' : 'sum', 'InvoiceDate' : 'first', 'CustomerID' : 'first'}

df = df.groupby('InvoiceNo').agg(operations)
```

```python
df.head()
```

```python
# The prediction will be based on year, create a column for year
df['year'] = df['InvoiceDate'].apply(lambda x: x.year)
```

```python
df.head()
```

```python
# Since date of transaction is also a source of feature
df['days_since'] = (pd.datetime(year=2010, month=12, day=31) - df['InvoiceDate']).apply(lambda x : x.days)
```

```python
# Calculate the minimum and maximum number of days and group by unique values
operations = {'revenue':'sum', 'days_since':['max', 'min', 'nunique'],}

x = df[df['year'] == 2010].groupby('CustomerID').agg(operations)
```

```python
x.head()
```

```python
# Column name rename
x.columns = [' '.join(col).strip() for col in x.columns.values]
```

```python
x.head()
```

```python
# Calculate average spend per order
x['avg_order_cost'] = x['revenue sum']/x['days_since nunique']
```

```python
y = df[df['year'] == 2011].groupby('CustomerID')['revenue'].sum()
```

```python
# Merge predictors and outcomes into a single dataframe
wrangled_df = pd.concat([x, y], axis=1)
wrangled_df.columns = ['2010 revenue',
                      'days_since_first_purchase',
                      'days_since_last_purchase',
                      'number_of_purchases',
                      'avg_order_cost',
                      '2011 revenue']
```

```python
wrangled_df.head()
```

```python
# There are so many values in the dataframe caused by some customers that were either actived in 2010 
# but not in 2011 or active in 2011 but not in 2010
wrangled_df = wrangled_df[~wrangled_df['2010 revenue'].isnull()]
wrangled_df = wrangled_df[~wrangled_df['2011 revenue'].isnull()]
```

```python
wrangled_df
```

```python
# Get rid of outliers
wrangled_df = wrangled_df[wrangled_df['2011 revenue'] < ((wrangled_df['2011 revenue'].median()) + 
                                                         wrangled_df['2011 revenue'].std()*3)]
wrangled_df = wrangled_df[wrangled_df['2010 revenue'] < ((wrangled_df['2010 revenue'].median()) + 
                                                         wrangled_df['2010 revenue'].std()*3)]
                                                         
```

```python
# Save the file to a new file
wrangled_df.to_csv('wrangled_transactions.csv')
```

### Building the Linear Regression Model

```python
df = pd.read_csv('wrangled_transactions.csv')
```

```python
df.corr()
```

```python
X = df[['2010 revenue',
      'days_since_last_purchase',
      'number_of_purchases',
      'avg_order_cost']]
```

```python
y = df['2011 revenue']
```

```python
# Split dataset to train and test data
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 100)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Get the model coeeficients
model.coef_
```

```python
# Check the intercepts
model.intercept_
```

```python
# Model equations
```

### Lets make a prediction

```python
single_customer = pd.DataFrame({
    '2010 revenue' : [1000],
    'days_since_last_purchase' : [20],
    'number_of_purchases' : [2],
    'avg_order_cost' : [500]
})
```

```python
model.predict(single_customer)
```

### Plot the model's prediction against true value

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(model.predict(X_test), y_test)
plt.xlim(0,10000)
plt.ylim(0,10000)
plt.plot([0, 10000], [0, 10000], 'k-', color = 'r')
plt.xlabel('Model Prediction')
plt.ylabel("True Value")
plt.show()
```

```python
from scipy.stats.stats import pearsonr

pearsonr(model.predict(X_test), y_test)
```
